<h3><center>State and Measurement model for cell ...</center></h3>

$\begin{array}{rclll}
&\dot{z{(t)}} &=& z{(t_0)} - \frac{1}{Q}\int_{t_0}^{t}\eta{(\tau)}i{(\tau)}d\tau &\text{(SoC
 equation in continuous time)}\\ \\
&\dot{i_R{(t)}}&=&-\frac{1}{R_{1}C_{1}}i_R{(t)}+\frac{1}{C_{1}}i{(t)} &\text{(Current\ in\ RC\ branches)}
\\ \\
&V{(t)}&=&OCV{(z{(t)})}-R_0 i{(t)} & {(R_{int}\ \text{model to compute cell OCV})}
\end{array}$

$\begin{array}{lcll}
\text{where :}\\ \\
&z{(t)} &:=& \text{State of charge at time instant t}&&\\ \\
&Q &:=& \text{Cell capacity}&&\\ \\ 
&\eta{(\tau)} &:=& \text{Coulombic efficiency at time instant t}&&\\ \\
&i{(\tau)} &:=& \text{Discharge current at time instant t}&&\\ \\
\end{array}$
<br><br>

$\text{The Enhanced Self Correcting Cell Model (ESC) :}$
$\begin{array}{rclll}
\quad \quad \quad\\
&z_{k+1}&=&z_k-\frac{\eta{_k}i_k\Delta t}{Q}&\mathrm{(SoC\ equation\ in\ discrete\ time)}\\ \\
&i_{R,k+1}&=&A_{RC}i_{R,k}+B_{RC}i_k&\mathrm{(Current\ in\ RC\ branches)}\\ \\
&h_{k+1}&=&{A_{H,k}}h_k+{(A_{H,k}-1)}sgn{(i_k)}&\mathrm{(hysteresis\ current\ equation)}\\
\end{array}$
<br><br>

$A_{H,k}=\exp\boldsymbol{{\left(-\left|\frac{{\eta{_k}i_k\gamma\Delta{t}}}{Q}\right|\right)}}$

$\begin{bmatrix}i_{R_1,k+1}\\i_{R_2,k+1}\\\vdots\\i_{R_n,k+1}\end{bmatrix}
=\underbrace{\begin{bmatrix}\exp{(\frac{-\Delta{t}}{R_1C_1})}&0&...\\ 0&\exp{(\frac{-\Delta{t}}{R_2C_2})}&...\\\vdots&\vdots&\ddots\end{bmatrix}}_{A_{RC}}
\begin{bmatrix}{i_{R_1,k}}\\{i_{R_2,k}}\\{\vdots}\end{bmatrix}
+\underbrace{\begin{bmatrix}{(1-\exp{(\frac{-\Delta{t}}{R_1C_1})})}\\{(1-\exp{(\frac{-\Delta{t}}{R_2C_2})})}\\{\vdots}\end{bmatrix}}_{B_{RC}}
i_k$

$\underbrace{\begin{bmatrix}z_{k+1}\\i_{R,k+1}\\h_{k+1}\end{bmatrix}}_{\large{x}\left[{k+1}\right]}
=\underbrace{\begin{bmatrix}1&0&0\\0&A_{RC}&0\\0&0&A_{H,k}\end{bmatrix}}_{\large{A\left[k\right]}}
\underbrace{\begin{bmatrix}z_k\\i_{R,k}\\h_k\end{bmatrix}}_{\large{x\left[{k}\right]}}+
\underbrace{\begin{bmatrix}\frac{-\eta{_k}\Delta{t}}{Q}&0\\B_{RC}&0\\0&A_{H,k}-1\end{bmatrix}}_{\large{B\left[k\right]}}
\underbrace{\begin{bmatrix}i_k\\sgn{(i_k)}\end{bmatrix}}_{\large{u\left[k\right]}}
\mathrm{(state\ space\ model\ )}$<br><br>

$\text{ESC ``state equation"}\quad x_{k+1}=A_kx_k+B_ku_k \quad\text{describes all dynamic effects}$<br><br>
$\boldsymbol{v_k=OCV{(z_k,T_k)}+M{_0}s_k+Mh_k-\sum_jR_ji_{R_j,k}-R_0i_k}$
